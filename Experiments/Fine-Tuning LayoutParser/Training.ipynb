```python
import pandas as pd
import numpy as np
import cv2
import random
import json
import pandas as pd
from tqdm import tqdm
import matplotlib.pyplot as plt
import os

from pycocotools.coco import COCO
import layoutparser as lp

plt.rcParams['figure.figsize'] = [10, 10]
```

```python
config_path = "/content/drive/MyDrive/Synthesis Project/Training/config.yaml"
model_path = "/content/drive/MyDrive/Synthesis Project/Training/model_0004999.pth"
```

```python
model = lp.Detectron2LayoutModel(config_path,
                                 model_path,
                                 extra_config= ["MODEL.ROI_HEADS.SCORE_THRESH_TEST", 0.8],
                                 label_map= {0:"TextRegion", 1:"ImageRegion", 2:"TableRegion", 3:"MathsRegion", 4:"SeparatorRegion", 5:"OtherRegion"})
```

```python
img_folder = "/content/drive/MyDrive/Synthesis Project/Sample Images/"


files = os.listdir(img_folder)
random_image = random.choice(files)
image_path = os.path.join(img_folder, random_image)

image = cv2.imread(image_path)
plt.imshow(image)
plt.show()
```

```python
layout_predicted = model.detect(image)
```

```python
color_map = {
    'TextRegion':   'red',
    'ImageRegion':  'blue',
    'TableRegion':   'green',
    'SeparatorRegion':  'yellow',
    'OtherRegion': 'pink',
}

layout = lp.draw_box(image,
              [b.set(id=f'{b.type}/{b.score:.2f}') for b in layout_predicted],
              color_map=color_map,
              show_element_id=True, id_font_size=10,
              id_text_background_color='black',
              id_text_color='white')

plt.imshow(layout)
plt.show()
```
